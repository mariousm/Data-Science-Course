# Carga de datos

```python
import pandas as pd
import numpy as np

pelis = pd.read_csv("data/movies.csv").drop(columns=["titulo"])
variable_objetivo = "ventas"

# hay peliculas sin la variable ventas
pelis = pelis[pelis[variable_objetivo].notnull()]

variables_independientes = pelis.drop(variable_objetivo, axis=1).columns


pelis.head()
```

```python
pelis.shape
```


Vamos a hacer codificacion one hot de las variables categ√≥ricas usando el OneHotEncoder de category_encoders:

```python
from category_encoders import OneHotEncoder


pelis = OneHotEncoder().fit_transform(pelis)
pelis.head()
```

```python
from sklearn.impute import SimpleImputer

X = SimpleImputer(strategy="mean").fit_transform(pelis.drop(variable_objetivo, axis=1))
y = pelis[variable_objetivo]
```

```python
variables_arbol = pelis.drop(variable_objetivo, axis=1).columns
```

```python
variables_arbol
```

# Crear un modelo √Årbol de Decision de profundidad m√°xima 5, entrenar y obtener grafo del √°rbol usando webgraphviz.

```python
from sklearn import tree

arbol = tree.DecisionTreeRegressor(max_depth=5).fit(X, y)
```

```python
tree.export_graphviz(arbol, out_file="arbol.dot", filled=True,
            feature_names=pelis.drop(variable_objetivo, axis=1).columns
)
```

# Variables m√°s importantes del modelo

```python
sorted(zip(arbol.feature_importances_,
           pelis.drop(variable_objetivo, axis=1).columns), reverse=True)
```
